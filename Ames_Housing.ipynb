**Question 1**: Download and load the Ames CSV file. Read the file into a Pandas DataFrame and display the first five rows.  

```python
import pandas as pd

df = pd.read_csv('AmesHousing.csv')

display(df.head())
```

**Question 2**:  Print the highest and lowest priced homes in Ames.

```python
max_sale_price = df['SalePrice'].max()
min_sale_price = df['SalePrice'].min()

print(f'The most expensive home in Ames is ${max_sale_price}')
print(f'The least expensive home in Ames is ${min_sale_price}')
```

**Question 3**:  How many categories does lot size have?

```python
shape_categories = df['Lot Shape'].nunique()

print(f'Lot Shape has {shape_categories} categories.')
```

**Question 4**:  What is the range of years for the house listings in the data?

```python
min_year_sold = df['Yr Sold'].min()
max_year_sold = df['Yr Sold'].max()

print(f'Houses in Ames have been sold between {min_year_sold} and {max_year_sold} for a total of {max_year_sold - min_year_sold} years.')
```

**Question 5**:  What is the total number of all homes sold in Ames for all years?

```python
count_sold = df['Yr Sold'].count()
print(f'{count_sold} homes have been sold between {min_year_sold} and {max_year_sold} in Ames.')
```

**Question 6**:  How many nulls or NA are there in the Lot Frontage column?

```python
null_lot_frontage = df['Lot Frontage'].isna().sum()

print(f'There are {null_lot_frontage} Null or NA values in the Lot Frontage column.')
```

**Question 7**:  How many items have a value in Alley and what is the total number of  NA in the column?

```python
count_alley_items = df['Alley'].count()
null_alley_items = df['Alley'].isna().sum()

print(f'There are {count_alley_items} Alley items and {null_alley_items} Null or NA items.')
```

**Question 8**:  What are all the homes in Gilbert with Yr Sold being 2009 and 2010?

```python
filter_df = df.query("Neighborhood == 'Gilbert' & (`Yr Sold` == 2009 | `Yr Sold` == 2010)")
#filter_df = df[(df['Neighborhood'] == 'Gilbert') & (df['Yr Sold'].isin([2009, 2010]))]

display(filter_df)
```

**Question 9**:  What is the count of homes sold in Gilbert with Yr Sold being 2010?

```python
count_gilbert_2010 = filter_df[filter_df['Yr Sold'] == 2010].shape[0]

print(f'{count_gilbert_2010} homes have been sold in Gilbert in 2010.')
```

**Question 10**:  Return the value of the third row, and fifth column.

```python
value = df.iloc[2, 4]

display(value)
```

**Question 11**:  Remake a new column from the SalePrice that includes data formatted with the $, and commas for the thousands/millions digits.

```python
df['SalePriceUSD'] = df['SalePrice'].apply(lambda x: "${:,.0f}".format(x))

display(df.head())
```

**Question 12**:   Output a dataframe with only SalePrice, Neighborhood, and Year Built.  Save this file in your git repository as “minimal_ames_housing.csv”,  Make sure you include the formatted SalePrice from the previous question.

```python
df_for_git = df[['Neighborhood', 'Year Built', 'SalePrice', 'SalePriceUSD']]

#Include index because original dataframe contained indicies
df_for_git.to_csv('minimal_ames_housing.csv', index=True)
```
